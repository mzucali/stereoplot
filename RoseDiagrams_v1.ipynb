```python
# prompt: i need a code for rose diagram also called petal diagram to plot data read from excel file containing columns with strike and dip

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming your excel file is named 'data.xlsx' and has columns 'strike' and 'dip'
# Replace 'data.xlsx' with your actual file name
#data = pd.read_excel('data.xlsx')

# Extract strike and dip values
#strike = data['strike'].values
#dip = data['dip'].values

strike = [12, 23, 34, 45, 56, 67,78]
dip = [2, 3,4 ,5,5,3,6]


# Convert strike to radians (0-2*pi)
strike_rad = np.deg2rad(strike)

# Create a rose diagram (petal diagram)
n_bins = 36  # Number of bins for the rose diagram
bin_width = 2 * np.pi / n_bins

# Calculate the frequencies of strike directions in each bin
hist, bin_edges = np.histogram(strike_rad, bins=n_bins, range=(0, 2 * np.pi))

# Plot the rose diagram
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
angles = np.deg2rad(np.linspace(0, 360, n_bins, endpoint=False))
ax.bar(angles, hist, width=bin_width, bottom=0.0, color='blue', alpha=0.7)

# Customize the plot
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_title('Rose Diagram of Strike Directions')
plt.show()
```
